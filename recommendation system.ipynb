```python
def convert3(obj):
    List=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:             #for 3 names of actors from column 
            List.append(i['name'])
            counter+=1
        else:
            break
    return List
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(obj):
    List=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            List.append(i['name'])
            break
    return List
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```
